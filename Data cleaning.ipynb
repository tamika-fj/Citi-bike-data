```python
import pandas as pd
from pathlib import Path
```

```python
#create path to csv file:

citibike= Path("Resources/JC-202307-citibike-tripdata.csv")
```

```python
bike_df = pd.read_csv(citibike)
```

```python
bike_df
```

```python
# Write the new data to a CSV file
output_file_path= Path('Resources/recent_bike_data.csv') 
bike_df.to_csv(output_file_path, index=False)
```

```python
#drop null values and reset index
bike_cleaned_df = bike_df.dropna()
bike_cleaned_df= bike_cleaned_df.reset_index(drop=True)
bike_cleaned_df
```

```python
# Write the new data to a CSV file
output_file_path= Path('Resources/recent_bike_data_cleaned.csv') 
bike_cleaned_df.to_csv(output_file_path, index=False)
```
